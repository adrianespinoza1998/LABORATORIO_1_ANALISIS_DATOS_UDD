```python
# Importamos las librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar estilo de gráficos
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6) # Tamaño por defecto para los gráficos
```

```python
df = pd.read_csv("paired_bladder_2022_clinical_data_clean.csv")
```

```python

# Mostrar las primeras filas de nuestro DataFrame
print("Primeras filas de los datos:")
print(df.head())

# Obtener un resumen estadístico de las variables numéricas
print("\nResumen estadístico de los datos:")
print(df.describe())

# Información general del DataFrame
print("\nInformación general del DataFrame:")
df.info()

df.columns
```

```python
df_reduced = df[['Overall Survival Status', 'TMB (nonsynonymous)', 'Cancer Type', 'Age at Which Sequencing was Reported (Years)', 'Smoker_never', 'Metastatic Site', 'Fraction Genome Altered']]

print("\nGenerando Pair Plot...")
sns.pairplot(df_reduced, hue='Overall Survival Status', diag_kind='kde') # 'hue' para diferenciar por tipo de cáncer
plt.suptitle('Pair Plot de Datos de Cáncer de Vejiga', y=1.02) # Título general
plt.show()
print("Pair Plot generado.")
```

```python
# Calculamos la matriz de correlación para las variables numéricas
matriz_correlacion = df_reduced.corr(numeric_only=True) # Solo variables numéricas

# Creamos un mapa de calor
plt.figure(figsize=(8, 6)) # Ajustamos el tamaño del gráfico
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Mapa de Calor de Correlación de Variables Numéricas', fontsize=16)
plt.show()
```
